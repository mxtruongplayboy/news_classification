```python
import pandas as pd
import os
import glob

# Đường dẫn tới thư mục chứa các file CSV
folder_path = r'E:\LEARN_5\XuLyNgonNguTuNhien\project\data_raw_2'

# Lấy danh sách tất cả các file CSV trong thư mục
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

# Đọc và gộp tất cả các file
df_list = [pd.read_csv(file) for file in csv_files]
merged_df = pd.concat(df_list, ignore_index=True)
```

```python
merged_df.head()
```

```python
# Thống kê các giá trị trùng nhau trong cột 'title'
duplicated_titles = merged_df['Title'].value_counts()
duplicated_titles = duplicated_titles[duplicated_titles > 1]

# Hiển thị kết quả
print("Các giá trị bị trùng trong cột 'title' và số lần xuất hiện:")
print(duplicated_titles)
```

```python
# Xác định những giá trị 'title' bị trùng
duplicated_titles = merged_df['Title'].duplicated(keep=False)

# Loại bỏ toàn bộ các hàng có giá trị 'title' bị trùng
clean_df = merged_df[~duplicated_titles]
```

```python
# Bước 1: Đọc dữ liệu từ file CSV
df = clean_df.copy()

# Bước 2: Giữ lại các cột cần thiết
df = df[['Title', 'Description', 'Content', 'Topic']]

# Bước 3: Loại bỏ giá trị thiếu
df = df.dropna(subset=['Title', 'Description', 'Content', 'Topic'])

df.head()
```

```python
df.info()
```

```python
# Hàm đọc từ viết tắt từ file txt
def load_abbreviations(file_path):
    abbreviations = {}
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            # Loại bỏ khoảng trắng và chia theo dấu ":"
            parts = line.strip().split(":")
            if len(parts) == 2:
                # Lưu từ viết tắt và từ đầy đủ vào dictionary mà không thay đổi gì
                abbreviations[parts[0].strip()] = parts[1].strip()  # Giữ nguyên phần từ đầy đủ
    return abbreviations

abbreviations = load_abbreviations(r'E:\LEARN_5\XuLyNgonNguTuNhien\project\crawl\acronym.txt')
```

```python
# Bước 4:  Hàm thay thế từ viết tắt trong văn bản
def replace_abbreviations_in_text(text):
    for abbr, full_text in abbreviations.items():
        text = re.sub(rf'\b{abbr}\b', full_text, text, flags=re.IGNORECASE)
    return text

df['Title'] = df['Title'].apply(replace_abbreviations_in_text)
df['Description'] = df['Description'].apply(replace_abbreviations_in_text)
df['Content'] = df['Content'].apply(replace_abbreviations_in_text)
```

```python
# Bước 5: Loại bỏ dấu câu và ký tự đặc biệt
def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text) 
    return text

df['Title'] = df['Title'].apply(clean_text)
df['Description'] = df['Description'].apply(clean_text)
df['Content'] = df['Content'].apply(clean_text)
```

```python
# Bước 6: Tải mô hình VnCoreNLP và load nó
model = py_vncorenlp.VnCoreNLP(save_dir='E:/LEARN_5/XuLyNgonNguTuNhien/vncorenlp')

def tokenize_text(text):
    # Tokenize using VnCoreNLP
    return model.annotate_text(text)
```

```python
# Bước 7: Tokenization (tách từ bằng py_vncorenlp)
df['Title_tokens'] = df['Title'].apply(tokenize_text)
df['Description_tokens'] = df['Description'].apply(tokenize_text)
df['Content_tokens'] = df['Content'].apply(tokenize_text)

df.head()
```

```python
# Đọc stopwords từ file
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        stopwords = f.read().splitlines()
    return stopwords

stopwords = load_stopwords('E:\\LEARN_5\\XuLyNgonNguTuNhien\\project\\crawl\\stopwords.txt')
```

```python
# copy bản sao df
df_copy = df.copy()
# Bước 8: Loại bỏ từ dừng (stopwords) sau khi tokenize và lấy wordForm
# Hàm lấy wordForm và chuyển thành chữ thường
import re

# Hàm lấy wordForm và chuyển thành chữ thường
def extract_wordforms(tokens_dict):
    # Lấy danh sách các token từ phần tử có key 0
    tokens = tokens_dict.get(0, [])
    # Lấy wordForm và chuyển thành chữ thường
    return [token['wordForm'].lower() for token in tokens if isinstance(token, dict) and token.get('wordForm')]

# Hàm loại bỏ stopwords
def remove_stopwords(wordforms, stopwords):
    # Loại bỏ các wordForm có trong stopwords
    return [word for word in wordforms if word not in stopwords]

# Hàm loại bỏ số
def remove_numbers(wordforms):
    # Loại bỏ các từ chứa số
    return [word for word in wordforms if not re.search(r'\d', word)]

# Hàm ghép lại thành câu
def join_to_sentence(wordforms):
    return ' '.join(wordforms)

# Áp dụng cho các cột trong DataFrame
df['Title_cleaned'] = df['Title_tokens'].apply(lambda x: join_to_sentence(remove_numbers(remove_stopwords(extract_wordforms(x), stopwords))))
df['Description_cleaned'] = df['Description_tokens'].apply(lambda x: join_to_sentence(remove_numbers(remove_stopwords(extract_wordforms(x), stopwords))))
df['Content_cleaned'] = df['Content_tokens'].apply(lambda x: join_to_sentence(remove_numbers(remove_stopwords(extract_wordforms(x), stopwords))))

# Kiểm tra kết quả
df.head()

```

```python
# Bước 9: Kết hợp các cột văn bản thành một cột duy nhất
df['text'] = df['Title_cleaned'] + " " + df['Description_cleaned'] + " " + df['Content_cleaned']

df.head()

# Bước 10: Lưu lại dữ liệu đã xử lý vào file CSV
df.to_csv(r'E:\LEARN_5\XuLyNgonNguTuNhien\project\data\cleaned_dulich.csv', index=False)
```
