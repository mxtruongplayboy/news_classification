```python
df = pd.read_csv(r'E:\LEARN_5\XuLyNgonNguTuNhien\vncorenlp\cleaned_news_data.csv') 
```

```python
df['content'] = df['text'].astype(str)
```

```python
word_list = ' '.join(df['content']).split()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

# Giả sử word_list là danh sách các từ trong văn bản của bạn
# Ví dụ: word_list = ['tin', 'tức', 'về', 'chính', 'trị', ...]

# Đếm tần suất từ
word_freq = Counter(word_list)

# Sắp xếp các từ theo tần suất giảm dần
sorted_word_freq = dict(sorted(word_freq.items(), key=lambda item: item[1], reverse=True))

# Lấy 30 từ xuất hiện nhiều nhất
top_30_words = dict(list(sorted_word_freq.items())[:30])

# Vẽ biểu đồ tần suất từ cho các từ phổ biến (30 từ nhiều nhất)
plt.figure(figsize=(10, 6))
plt.bar(top_30_words.keys(), top_30_words.values())
plt.xticks(rotation=90)
plt.xlabel('Từ')
plt.ylabel('Tần suất')
plt.title(f'Biểu đồ tần suất 30 từ phổ biến nhất')
plt.show()

# Tạo đám mây từ cho tất cả các từ trong word_freq (không lọc theo ngưỡng)
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Hiển thị đám mây từ
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Đám mây từ (Tất cả từ trong văn bản)')
plt.show()
```
